```python
import numpy as np
import pandas as pd
my_df = pd.read_csv('dog_info.csv')
print(my_df)
```

```python
my_df['color'].value_counts() # unique value ko define krta hai 
# ye zayada value nicha ki tarf jata hain mera 1/1 is lia hai 
# ka mera pass sirf 1 hai   
my_df['color'].value_counts(ascending = True) # ye os ka opposite hai
my_df['pentname'].value_counts(dropna = False) # ye nun value ko be likhta hai or pentname cgpt ki mistake hai csv file
# cgpt sa le hai 
my_df['pentname'].value_counts(normalize= True) # % nikal kr deta hai ka is name ka ktna % pets hain
my_df['pentname'].value_counts()['Charlie'] # is ma ye bata raha Charlie name ki ktna pets hain
my_df.groupby('color').size() # is me ye print ga ka color group me ktna pets hain like white/orange , white/black 
# jin me ye 2 color mix ho
```

# Pandas series

```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

```python
my_series = [1, 88, 21, 233]
my_new = pd.Series(my_series) 
print(my_new)
```

```python
my_index = ['A','B','C','D'] # ye sary index hain
my_new = pd.Series (my_series,my_index) #pd.Series() Pandas ka function hai, jo ek Series object banata hai.
#A Pandas Series ek 1-dimensional labeled array hota hai (like a column of data).
print(my_new)
my_new['B'] # is me B mane index hai is me jo be number ye value ho gi wo print ho jya gi
```

# Key Value Dictionary

```python
cars = {"Honda":21 , "Tesla":30 , "BMW" : 10 }
my_cars = pd.Series(cars)
print(my_cars)
cars["Honda"]
```

# DataFrame

```python
my_data = randn(4,3) # 4 rows hain 3 column
my_row = ['A','B','C','D']
my_col = ['Monday','Tuesday','Wednesday']

my_df = pd.DataFrame(my_data,my_row,my_col)
print(my_df)
```
